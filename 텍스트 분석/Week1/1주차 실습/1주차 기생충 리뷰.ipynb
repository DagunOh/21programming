```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url='http://www.cgv.co.kr/search/review.aspx?query=%ea%b8%b0%ec%83%9d%ec%b6%a9'
```

```python
res=requests.get(url)
```

```python
source = BeautifulSoup(res.text, 'html.parser')
```

```python
review_only=source.find_all('div',{'class':"box-contents"})
```

```python
for review in review_only:
    print(review.get_text().strip())
```
